```python
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lis=os.listdir('./dogs/')
```

```python
l=[]
for i in lis:
    l.append(('/home/murky/Abhishek/hackathon/SNU/dogs/'+i))
# print(l)
    
```

```python
def resizing(img):
    img=cv2.imread(img,1)
    r = 100.0 / img.shape[1]
    dim = (100, int(img.shape[0] * r))
    img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    img = cv2.resize(img, (64, 64))
#     iINTER_AREAmg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
    
```

```python
for i, img in enumerate(l):
    image = resizing(img)
    cv2.imwrite("./dogr/dogs-100"+str(i)+".jpg", image)
```
